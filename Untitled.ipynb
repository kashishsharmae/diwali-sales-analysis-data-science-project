```python
#  1. Import Required Librararies
# Importing essential libraries for data manipulation and visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Configure plot aesthetics for better visuals
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
#  2. Load Dataset 
# Load the synthetic dataset into a DataFrame
df = pd.read_csv('diwali_sales_unique.csv')
print("Data loaded successfully. Shape:", df.shape)
```

```python
#  3. Data Overview
# Preview the first few rows and basic structure of the dataset
print(df.head())               # Displays first 5 records
print(df.info())               # Shows data types and null values
print(df.describe())          # Summary statistics of numerical columns
```

```python
#  4. Data Cleaning 
# Check for any missing values
print("Missing values per column:\n", df.isnull().sum())
```

```python
# Remove duplicate entries if any
df.drop_duplicates(inplace=True)
```

```python
# Ensure numerical fields are correctly typed
# Convert Amount, Quantity, and Price_per_Unit to integers
df['Amount'] = df['Amount'].astype(int)
df['Quantity'] = df['Quantity'].astype(int)
df['Price_per_Unit'] = df['Price_per_Unit'].astype(int)
```

```python
#  5. Exploratory Data Analysis (EDA) 
# 5.1 Total sales by state
# Group by State and sum the sales Amount
state_sales = df.groupby('State')['Amount'].sum().sort_values(ascending=False)
state_sales.plot(kind='bar', color='coral', title='Total Sales by State')
plt.xlabel('State')
plt.ylabel('Total Sales Amount')
plt.tight_layout()
plt.show()
```

```python
# 5.2 Gender-wise customer count
# Count plot showing distribution of Male and Female customers
sns.countplot(data=df, x='Gender', palette='Set2')
plt.title('Customer Gender Distribution')
plt.show()
```

```python
# 5.3 Gender-wise sales amount
# Total sales amount contributed by each gender
gender_sales = df.groupby('Gender')['Amount'].sum()
gender_sales.plot(kind='bar', color=['blue', 'pink'], title='Sales by Gender')
plt.ylabel('Total Sales')
plt.show()
```

```python
# 5.4 Age group analysis by gender
# Count of each age group broken down by gender
sns.countplot(data=df, x='Age_Group', hue='Gender', palette='husl')
plt.title('Age Group Distribution by Gender')
plt.show()
```

```python
# 5.5 Sales by Product Category
# Total sales amount for each product category
cat_sales = df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=True)
cat_sales.plot(kind='barh', color='mediumseagreen', title='Sales by Product Category')
plt.xlabel('Total Sales')
plt.show()
```
